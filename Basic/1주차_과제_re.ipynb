```python
array1 = np.array([1, 2, 3])
array2 = np.array([[1, 2, 3], [4, 5, 6]])

# 넘파이 array데이터 타입 --> numpy.ndarray
# array 형태 출력 / array_name.shape
print('type:', type(array1))
print('shape:', array1.shape)
print('shape:', array2.shape)
```

2. ndarray 생성하기

np.arrange(10)  
np.zeros((3, 2), dtype='int32')  
np.ones((3, 2), dtype='int32')

```python
# 1차원 배열 n개 원소 생성 / np.arange(n)
array1 = np.arange(10)
print(array1)

# np.zeros((), dtype = '?)' / 원소 0인 배열 생성 
# np.ones((,), dtype = '?') / 원소 1인 배열 생성
array2 = np.zeros((3, 3, 3), dtype='int32')
array3 = np.ones((2, 3))
print('array2 = \n', array2, '\n\n','array3 = \n', array3)
```

```python
# -1은 크기에 알맞게 값을 조정하하는 뜻임
array_re = array2.reshape(-1, 3)
print(array_re)
```

5. ndarray 인덱싱
- 단일값추출
- 슬라이싱
- 팬시 인덱싱
- 불린 인덱싱

```python
# 특이한거만 다루겠음 
array1 = np.arange(1, 28)
array1 = array1.reshape(3, 3, 3)
print(array1, '\n')

# 팬시 인덱싱(사실 잘모르게ㅖㅆ음)
print(array1[[1, 2], [1, 2]], '\n')

# 불린 인덱싱
print(array1[array1 > 9])
```

6. ndarray 정렬  
np.sort()  
np.argsort()

```python
# np.sort / 정렬 호출, 기존 array에는 변화 x
messy_array = np.array([1, 3, 2, 4])
clean_array = np.sort(messy_array)
print(clean_array)
print(messy_array, '\n')

# array_name.sort()/ 정렬 호출 x, 기존 array 정렬 반환
clean_array_2 = messy_array.sort()
print(clean_array_2)
print(messy_array)
```

#### axis 간단 정리  
axis == 축, axis가 작을 수록 더 작은 차원의 축 지정 
  
ex) 2차원 ndarray  
[[1, 2], [3, 4]]  
axis = 0 | [1, 2] --> [3, 4] 바깥쪽 괄호, row를 나타냄  
axis = 1 | 1 --> 2, 3 --> 4 안쪽 괄호, column을 나타냄  

```python
# np.argsort(ndarray) / 각 index의 크기 순위를 매김
messy_array = np.array([1, 3, 2, 4])
arg_array = np.argsort(messy_array)
print(arg_array)

```

```python
# argsort 활용 예시
name = np.array(['kim', 'lee', 'chu'])
score = np.array([30, 20, 100])
ranking = np.argsort(score)
print(name[ranking])
# wow
```

#### 7. dot production / transpose
np.dot(A, B) --> 행렬 곱
np.transpose(A) --> 전치행렬

```python
A = np.array([[1, 2, 3], 
              [4, 5, 6]])
B = np.array([[1, 2], 
              [3, 4], 
              [5, 6]])
dot_product = np.dot(A, B)
transpose_matrix = np.transpose(A)
print(dot_product, '\n')
print(transpose_matrix)
```
