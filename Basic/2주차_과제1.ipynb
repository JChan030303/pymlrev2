#### 2. DF에서의 리스트, 딕셔너리, ndarray

```python
import numpy as np
lst = [1, 2, 3]
ndarr = np.array(lst)

# 1. list --> df
df_lst = pd.DataFrame(lst, columns = ['column1'])
print(df_lst, '\n')

# 2. ndarray --> df
df_ndarr = pd.DataFrame(ndarr, columns = ['columns2'])
print(df_ndarr)

# +)column 여러개
col_name = ['col1', 'col2', 'col3']
lst2 = [[1, 2, 3], [4, 5, 6]]
df_lst2 = pd.DataFrame(lst2, columns = col_name)
print(df_lst2, '\n')

# 3. dict --> df
# key --> 칼럼명, value = 해당 칼럼의 데이터
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print(df_dict)

#---------------------------------------------------------

# 4. df --> ndarray
arr = df_dict.values
print(arr, '\n')

# 5. df --> list
# 정확히 하면 df --> ndarray --> list
# np.tolist()
lst = df_dict.values.tolist()
print(lst)

# 6. df --> dict
dict = df_dict.to_dict('list') # column --> key, column의 value(리스트 형태) --> value
print(dict)
```

#### 결손 데이터 처리 


```python
# 1. 결손 데이터 확인
# df_name.isna()
print(titanic_df.isna().sum(), '\n') # 여기에 sum하는데 왜 이렇게 되냐 신기하노

# 2. 결손 데이터 채우기 
# df_name.fillna()
# 'Cabin' 열
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df[titanic_df['Cabin'] == 'C000'].head(5)

# 'Age' 열
# 남자na --> 남자 중간값 / 여자na --> 여자 중간값
# 내가 따로 한거 ★
male_age_mid = titanic_df.loc[titanic_df['Sex'] == 'male', ['Age']].median()
female_age_mid = titanic_df.loc[titanic_df['Sex'] == 'female', ['Age']].median()
titanic_df.loc[titanic_df['Sex'] == 'male', ['Age']] = titanic_df.loc[titanic_df['Sex'] == 'male', ['Age']].fillna(male_age_mid)
titanic_df.loc[titanic_df['Sex'] == 'female', ['Age']] = titanic_df.loc[titanic_df['Sex'] == 'female', ['Age']].fillna(female_age_mid)
print(titanic_df.loc[:, ['Age', 'Sex']].head(15))
```
