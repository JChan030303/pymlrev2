#### 2. DF에서의 리스트, 딕셔너리, ndarray

```python
import numpy as np
lst = [1, 2, 3]
ndarr = np.array(lst)

# 1. list --> df
df_lst = pd.DataFrame(lst, columns = ['column1'])
print(df_lst, '\n')

# 2. ndarray --> df
df_ndarr = pd.DataFrame(ndarr, columns = ['columns2'])
print(df_ndarr)

# +)column 여러개
col_name = ['col1', 'col2', 'col3']
lst2 = [[1, 2, 3], [4, 5, 6]]
df_lst2 = pd.DataFrame(lst2, columns = col_name)
print(df_lst2, '\n')

# 3. dict --> df
# key --> 칼럼명, value = 해당 칼럼의 데이터
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print(df_dict)

#---------------------------------------------------------

# 4. df --> ndarray
arr = df_dict.values
print(arr, '\n')

# 5. df --> list
# 정확히 하면 df --> ndarray --> list
# np.tolist()
lst = df_dict.values.tolist()
print(lst)

# 6. df --> dict
dict = df_dict.to_dict('list') # dict 내부 value값을 list로 리턴
print(dict)
```
